```python
import datetime
import time
import re
import string
import pygtrie
```

```python
#Python doesnot accept the date with 0
#Therefore this function eliminates 0 from 01 and returns 1 instead
def delete0 (number):
    if number[0] ==0:
        number = number[1]
        return int(number)
    else:
        return int(number)
#reads the lines from the sample data file
f = open("AOL.txt.sorted","r",encoding='utf-8')
lines = f.readlines()
f.close()

f = open("AOL.txt.normalized", "w+",encoding='utf-8')
for line in lines:
    #NORMALIZATION
    stringToNormalized = line.split("\t")[1].upper()
    stringToNormalized = re.sub( 'WWW', '', stringToNormalized )
    stringToNormalized = re.sub( 'COM', '', stringToNormalized )
    stringToNormalized = "".join((char for char in stringToNormalized if char not in string.punctuation))
    if not stringToNormalized:
        stringToNormalized = "NULL"
    stringToNormalized = re.sub( ' +', ' ', stringToNormalized )

    #CONVERSION TO TIMESTAMP
    #Gets the date and time info
    dateAndTime =line.split("\t")[2]
    
    #Deals with the "date" part
    date = dateAndTime.split(" ")[0]
    dates = date.split("-")
    j =0
    dates_int = [0,0,0]
    for i in dates:
        dates_int[j] = delete0(dates[j])
        j = j+1
        
    #Deals with the "time" part
    timeT = dateAndTime.split(" ")[1]
    timesT = timeT.split(":")
    j =0
    timesT_int = [0,0,0]
    for i in timesT:
        timesT_int[j] = delete0(timesT[j])
        j = j+1
   
    year   = dates_int[0]
    month  = dates_int[1]
    day    = dates_int[2]
    hour   = timesT_int[0]
    minute = timesT_int[1]
    second = timesT_int[2]
    
    #Converts to the time stamp
    dt = datetime.datetime(year, month, day, hour, minute, second)
    unixtime = int(time.mktime(dt.timetuple()))
    
    #Saves the data to the file
    if(stringToNormalized != "NULL"):
        f.write(line.split("\t")[0] +"\t"+ stringToNormalized + "\t"+ str(unixtime) +"\t\n")
f.close()
```

```python
#CREATING LEXICON
#opens the normalized file to read the searched terms
f = open("AOL.txt.normalized","r")
lines = f.readlines()
f.close()

#creates dictionary
lexicon = dict()

#iterates through the lines
for line in lines:  
    terms = line.split("\t")[1]            #gets the searched part from line
    term = terms.split(" ")                #splits the line into the terms by spliting according to the space
    iteration = len(term)                  #gets the number of terms in search
    for i in range(iteration):             
        if term[i] not in lexicon:         #checks if the term is in the dictionary
            lexicon[term[i]] = [1,1]       #if not updates the dictionary with the new term
        else:
            lexicon[term[i]][1] = lexicon[term[i]][1] + 1 #updates the number of times that the term has used

#Creating TRIE to save the terms with their ids
i = 0
trie=pygtrie.StringTrie()
f = open("AOL.sorted.terms", "w+")
for key in lexicon:
    i = i+1
    trie[key] = i
    f.write(key + "\t" + str(lexicon[key][0]) + "\t" + str(lexicon[key][1])+ " " + "\n")
f.close()

f = open("AOL.txt.normalized", "r")
lines = f.readlines()
f.close()

#Saves the terms in their numerical order
f = open("AOL.txt.numerical", "w+")
for line in lines:
    lineArray = line.split("\t")
    userID = lineArray[0]
    timeStamp = lineArray[2]
    searchQuery = lineArray[1]
    term = searchQuery.split(" ")
    iteration = len(term)
    listTemp = []
    for i in range(iteration):
        listTemp.append(str(trie[term[i]]))              
    numSearch = str(' '.join(listTemp))
    f.write(userID + "\t" + timeStamp + "\t" + numSearch + "\n")
f.close()
```
